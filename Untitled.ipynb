# Imports

```python
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
```

# Load Data

```python
sentiment = pd.read_csv("fear_greed_index.csv")
sentiment
```

```python
trades = pd.read_csv("historical_data.csv")
trades
```

# Basic Checks

```python
print("Trade Shape",trades.shape)
print("Sentiment Shape",sentiment.shape)
```

```python
print("Trade Missing Value\n",trades.isnull().sum())
print("--------------------------------------------------")
print("Sentiment Missing Value\n",sentiment.isnull().sum())
```

```python
print("Trade Duplicate Value",trades.duplicated().sum())
print("Sentiment Duplicated Value",sentiment.duplicated().sum())
```

```python
print(trades.info())
print("--------------------------------------------------------------")
print(sentiment.info())
```

# Convert Timestamps

```python
# Trades timestamp fix (IMPORTANT)
trades['Timestamp IST'] = pd.to_datetime(trades['Timestamp IST'], dayfirst=True)
trades['Timestamp'] = pd.to_datetime(trades['Timestamp'], unit='s')

# Sentiment date
sentiment['timestamp'] = pd.to_datetime(sentiment['timestamp'], unit='s')
sentiment['date'] = pd.to_datetime(sentiment['date'])

print(trades.info())
print("--------------------------------------------------------------")
print(sentiment.info())
```

# Convert TradeID DataType

```python
trades['Trade ID'] = trades['Trade ID'].astype('int64')
trades.info()
```

# Daily Metrics

```python
daily = trades.groupby('Timestamp IST').agg(
    total_pnl=('Closed PnL', 'sum'),
    trades_per_day=('Trade ID', 'count'),
    avg_trade_size=('Size USD', 'mean')
).reset_index()

daily.head()
```

# Win Rate

```python
trades['win'] = np.where(trades['Closed PnL'] > 0, 1, 0)

win_rate = trades.groupby('Timestamp IST').agg(
    win_rate=('win', 'mean')
).reset_index()

daily = daily.merge(win_rate, on='Timestamp IST', how='left')

daily.head()
```

# Long/Short Ratio

```python
ls = trades.groupby(['Timestamp IST', 'Side']).size().unstack(fill_value=0)

ls['long_short_ratio'] = ls.get('BUY', 0) / ls.get('SELL', 1)

ls = ls.reset_index()[['Timestamp IST', 'long_short_ratio']]

daily = daily.merge(ls, on='Timestamp IST', how='left')

daily.head()
```

# Merge with Sentiment

```python
daily.rename(columns = {'Timestamp IST':'date'}, inplace = True)
```

```python
merged = pd.merge(
    daily,
    sentiment[['date', 'classification']],
    on='date',
    how='inner'
)

merged.head()
```

# Analysis Table

```python
summary = merged.groupby('classification').agg(
    avg_pnl=('total_pnl', 'mean'),
    avg_trades=('trades_per_day', 'mean'),
    avg_size=('avg_trade_size', 'mean'),
    avg_win_rate=('win_rate', 'mean'),
    avg_ls_ratio=('long_short_ratio', 'mean')
).reset_index()

summary
```

# Visualization

```python
# PnL
plt.figure()
plt.bar(summary['classification'], summary['avg_pnl'])
plt.title("Average PnL by Sentiment")
plt.xticks(rotation=45)
plt.show()

# Trades
plt.figure()
plt.bar(summary['classification'], summary['avg_trades'])
plt.title("Trades per Day by Sentiment")
plt.xticks(rotation=45)
plt.show()

# Trade Size
plt.figure()
plt.bar(summary['classification'], summary['avg_size'])
plt.title("Average Trade Size by Sentiment")
plt.xticks(rotation=45)
plt.show()
```

# Segmentation

```python
# Activity segment
activity = trades.groupby('Account').agg(
    total_trades=('Trade ID', 'count')
).reset_index()

threshold = activity['total_trades'].median()

activity['segment'] = np.where(
    activity['total_trades'] > threshold,
    'High Frequency',
    'Low Frequency'
)

activity.head()
```

### Key Insights

1. PnL is higher during Fear compared to Greed.
2. Trade frequency increases significantly during Fear.
3. Traders reduce position size during Extreme Fear.

### Strategy Ideas

1. During Fear → Increase trade frequency but control position size.
2. During Greed → Reduce trading and avoid large positions.
